```python
from ggplot import *
print(mtcars.head())
```

```python
p=ggplot(mtcars,aes(x='wt',y='hp'))+geom_line(color='blue',alpha=0.5)+geom_point(color='red',size=40,alpha=0.5)
print(p)
```

```python
p=ggplot(mtcars,aes(x='wt',y='hp',color='factor(gear)'))+geom_line()
print(p)
```

```python
p=ggplot(mtcars,aes(x='wt',y='hp',color='factor(gear)'))+geom_line()+facet_grid('factor(gear)')+xlim(2,6)
print(p)
```
