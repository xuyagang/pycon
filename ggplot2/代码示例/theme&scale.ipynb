```python
from ggplot import *
df=diamonds.sample(n=2000)
print(df.head())
```

```python
p=ggplot(aes('carat', 'price'), data=df) + geom_point()
print(p)
```

```python
p=ggplot(aes(x='carat', y='price', color='clarity'), data=df) +geom_point() 
print(p)
```

```python
p=ggplot(aes(x='carat', y='price', color='clarity'), data=df) +geom_point() +scale_color_brewer(type='qual')
#qual:
#div:
#seq:
#
print(p)
```

```python
p=ggplot(aes(x='carat', y='price', color='clarity'), data=df) +geom_point() +scale_color_brewer(type='seq', palette='Reds')
#
print(p)
```

```python
p=ggplot(aes(x='carat', y='price', color='clarity'), data=df) +geom_point() +scale_color_brewer(type='seq', palette='Greens')
#
print(p)
```

```python
p=ggplot(aes(x='carat', y='price', color='clarity'), data=df) +geom_point() +scale_color_brewer(type='div', palette=2)
#
print(p)
```

```python
p=ggplot(aes(x='carat', y='price', color='clarity'), data=df) +geom_point() +scale_color_brewer(type='seq', palette='Reds')+scale_x_log()
#
print(p)
```

```python
p=ggplot(aes(x='carat', y='price', color='clarity'), data=df) +geom_point() +scale_color_brewer(type='seq',\
                                                                                                palette='Greens')+labs(x='my carat',y='my price',\
                                                                                                                                   title='my point')
#
print(p)
```
