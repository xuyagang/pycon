```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy'))+geom_point()  #缺乏美学图层
print(p)  
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy'))+geom_point()+aes(colour='factor(mpg.year)')  
#
#python不支持aes的叠加，这个颜色必须放在数据层里
#
print(p) 
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy',color='factor(year)'))+geom_point()  #这样子就画出按年分类的散点图了

print(p) 
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy',color='factor(year)'))+geom_point(size=40)  #这样子就画出按年分类的散点图了
#geom_point中可以添加尺寸，颜色。注意在数据层和在几何层颜色的意义是不同的

print(p) 
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy',color='factor(year)'))+geom_point(size=40,color='red')  
#因为有一个图层的概念，所以后面的层会将下面的层给掩盖掉，这里颜色都变成了红色
#颜色在数据层是某个维度用颜色分类，在几何图形层是画什么颜色的意思
print(p) 
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy'))+geom_point(colour='black', size = 100) +geom_point(size=40,color='pink')
print(p) 
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy'))+geom_point(size=40,color='pink')+geom_point(colour='black', size = 100)
#跟上图做个对比，看一下为什么这样子？
#没错，就是因为图层是叠加的，后面的几何层覆盖前面的层。
print(p) 
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy',color='factor(year)'))+geom_point(size=40,alpha=0.2) 
#在几何层用alpha来显示透明度，是一个0-1的数，为1时最大，表示完全不透明

print(p) 
```

```python
from ggplot import *
p=ggplot(mpg,aes(x='cty',y='hwy',color='factor(year)'))+geom_point(size=40,alpha=0.5) 
#在几何层用alpha来显示透明度，是一个0-1的数，为1时最大，表示完全不透明

print(p) 
```
