```python
plot1 = ggplot(mtcars, aes('wt', 'mpg')) + geom_point()
print(plot1)
```

```python
ggplot(mtcars,aes('wt','mpg',color='cyl'))+geom_point()
```

```python
# 将颜色类别离散化
ggplot(mtcars,aes('wt','mpg',color='factor(cyl)'))+geom_point()
```

```python
ggplot(mtcars,aes('wt','mpg'))+geom_point(color='red',alpha=0.6)
```

### alpha 透明度用在数据密集的情形

```python
ggplot(diamonds,aes('carat', 'price', shape='cut', color='cut')) + geom_point(alpha=0.4)
```

### jitter 用于数据密集的情况
#### 类似于geom_point
Same as geom_point but with randomness added so you can see the points better

```python
ggplot(diamonds,aes('carat', 'price', shape='cut', color='cut')) + geom_jitter()
```

### 抽样 和 切片

```python
# 使用 frac 按比例抽取
df1 = diamonds.sample(frac=0.03)
# 随机抽取行
df2 = diamonds.sample(n=3000)
```

```python
plot1 =  ggplot(df1,aes('carat', 'price', shape='cut', color='cut')) + geom_point()
plot2 = ggplot(df2,aes('carat', 'price', shape='cut', color='cut')) + geom_point()

print(plot1)
print(plot2)
```

### 两层的嵌套

```python
ggplot(df1,aes('carat', 'price')) + geom_point(color='red', size=100) +\
geom_point(color='white', size=30)
```

### 分面 facet_wrap  & fact_grid

```python
ggplot(mtcars,aes('wt','mpg')) + geom_point() + facet_grid('cyl', scales='free')
```

### 分组

```python
ggplot(mtcars, aes('wt','mpg',color='factor(cyl)')) + geom_point()
```

### 美化

```python
ggplot(mtcars, aes('wt','mpg',color='factor(cyl)')) + geom_point() +\
labs(x='X', y='Y',title='我的标题') + xlim(2,4) + ylim(10,25)
```
