### alpha 

```python
ggplot(diamonds,aes('carat', 'price', shape='cut', color='cut')) + geom_point(alpha=0.4)
```

### jitter 
#### geom_point
Same as geom_point but with randomness added so you can see the points better

```python
ggplot(diamonds,aes('carat', 'price', shape='cut', color='cut')) + geom_jitter()
```

###   

```python
#  frac 
df1 = diamonds.sample(frac=0.03)
# 
df2 = diamonds.sample(n=3000)
```

```python
plot1 =  ggplot(df1,aes('carat', 'price', shape='cut', color='cut')) + geom_point()
plot2 = ggplot(df2,aes('carat', 'price', shape='cut', color='cut')) + geom_point()

print(plot1)
print(plot2)
```
