```python
from PIL import Image,ImageFilter
%matplotlib inline
import matplotlib.pyplot as plt
im = Image.open('./img/01.png')
plt.imshow(im)
```

#### blur    模糊

```python
blur_im = im.filter(ImageFilter.BLUR)
plt.imshow(blur_im)
```

#### contour    轮廓

```python
contour_im = im.filter(ImageFilter.CONTOUR)
plt.imshow(contour_im)
```

#### detail 细节

```python
detail_img = blur_im.filter(ImageFilter.DETAIL)
plt.imshow(detail_img)
```

#### edge_enhance 边界增强

```python
ege_img = im.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(ege_img)
```

#### edge_enhance_more

```python
eem_img = im.filter(ImageFilter.EDGE_ENHANCE_MORE)
plt.imshow(eem_img)
```

#### EMBOSS 雕刻、装饰

```python
em_img = im.filter(ImageFilter.EMBOSS)
plt.imshow(em_img)
```

#### find_edge

```python
# im = Image.open('./img/01.png')
fe_img = im.filter(ImageFilter.FIND_EDGES)
bass = Image.new('RGB', (fe_img.size[0],fe_img.size[1]))
bass.paste(fe_img)
plt.imshow(bass)
# fe_img.show()
```

#### SMOOTH

```python
sm_img = im.filter(ImageFilter.SMOOTH)
plt.imshow(sm_img)
```

#### SMOOTH_MORE

```python
sm_img = im.filter(ImageFilter.SMOOTH_MORE)
plt.imshow(sm_img)
```

#### SHARPEN

```python
sm_img = im.filter(ImageFilter.SHARPEN)
plt.imshow(sm_img)
```

#### GaussianBlur
模糊半径：radius

```python
sm_img = im.filter(ImageFilter.GaussianBlur(radius=5))
plt.imshow(sm_img)
```

#### UnsharpMask  反锐化掩码滤波
- radius：模糊半径
- percent：反锐化强度（百分比）
- threshold：被锐化的最小亮度

```python
sm_img = im.filter(ImageFilter.UnsharpMask(radius=2,percent=150,threshold=3))
plt.imshow(sm_img)
```

#### Kernel  卷积核滤波
- size：核的大小（width, height）
- kernel：核权值序列如3*3的为(1, 2, 2, 2, 0, 0, 1, 2, 2)
- scale：缩放因子
- offset：偏移量（使用的话，则将该值加到缩放后的结果上）

```python
k_img = im.filter(ImageFilter.Kernel((3, 3), (1, 2, 2, 2, 0, 0, 1, 2, 2), scale=None, offset=10))
plt.imshow(k_img)
```
