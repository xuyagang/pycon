```python
from PIL import Image,ImageFilter
%matplotlib inline
import matplotlib.pyplot as plt
im = Image.open('./img/01.png')
plt.imshow(im)
```

#### blur    模糊

```python
blur_im = im.filter(ImageFilter.BLUR)
plt.imshow(blur_im)
```

#### contour    轮廓

```python
contour_im = im.filter(ImageFilter.CONTOUR)
plt.imshow(contour_im)
```

#### detail 细节

```python
detail_img = blur_im.filter(ImageFilter.DETAIL)
plt.imshow(detail_img)
```

#### edge_enhance 边界增强

```python
ege_img = im.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(ege_img)
```

#### edge_enhance_more

```python
eem_img = im.filter(ImageFilter.EDGE_ENHANCE_MORE)
plt.imshow(eem_img)
```

#### EMBOSS 雕刻、装饰

```python
em_img = im.filter(ImageFilter.EMBOSS)
plt.imshow(em_img)
```

#### find_edge

```python
# im = Image.open('./img/01.png')
fe_img = im.filter(ImageFilter.FIND_EDGES)
bass = Image.new('RGB', (fe_img.size[0],fe_img.size[1]))
bass.paste(fe_img)
plt.imshow(bass)
# fe_img.show()
```
