```python
from PIL import Image,ImageFilter
%matplotlib inline
import matplotlib.pyplot as plt
im = Image.open('./img/02.jpg')
plt.imshow(im)
```

#### blur    模糊

```python
blur_im = im.filter(ImageFilter.BLUR)
plt.imshow(blur_im)
```

#### contour    轮廓

```python
contour_im = im.filter(ImageFilter.CONTOUR)
plt.imshow(contour_im)
```
