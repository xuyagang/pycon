### PIL库主要用途
- 图像存储
    - PIL适合图像归档和图像批量处理，你可以使用它建立缩略图，转换格式，打印图片等。现在的版本可以识别和读取大量的图片格式，写入常用的转换和表示格式
- 图像显示
    - 支持多个其他工具包的展示，提供了show()方法，可以保存图像到磁盘并调用外显示
- 图像处理（重点）
    - 包含了基本的图像处理功能，包括点操作，使用内置卷积内核过滤，色彩空间转换。支持更改图像大小、旋转、任意仿射变换

#### 图片属性
成功加载后会返回一个 Image 对象，可观察以下属性：
- format： 判断图片来源，即ppm/png等 (若不是从文件读取则返回None)；
- size： 查看宽度和高度(二元组tuple，单位是px)；
- mode： 定义图像波段数与名称(像素类型和深度)，常见有 L 灰度图像，RGP 真彩色图像， CMYK 出版图像


```python
im = Image.open('./img/01.png')
print(im.format,im.size,im.mode)
```

### 图像处理

#### 剪切、粘贴、合并
- crop()：从图像中提取子矩形。box坐标分别是(左，上，右，下)的位置
- paste()：粘贴。粘贴区域是大小必须完全匹配，但模式不需要匹配(模式在粘贴时会自动转换)。粘贴时可以选择透明度，0 对应完全透明，255对应不透明(默认)
- split()：用于对图像进行波段拆分，如针对RGB图像，拆分后将生成3幅单波段图像，对单波段图像则返回本身
- merge()：用于将多个波段图像合并成新图像

```python
box = [100,100,400,400]
im = Image.open('./img/01.png')
# 超原图范围的用空白填充
region = im.crop(box)
plt.imshow(region)
```

```python
im.paste(region,[0,20,region.size[0],region.size[1]+20])
plt.imshow(im)
```

```python
a,b,c,d = im.split()
plt.imshow(a)
```

```python
plt.imshow(b)
```

```python
plt.imshow(c)
```

```python
plt.imshow(d)
```

```python
im = Image.merge('RGB',[a,c,d])
plt.imshow(im)
```

#### 模式转换
convert()：对图像进行不同颜色模型的转换，如常用的 L 与 RGB 之间的转换
- 1：1位像素，表示黑和白，但是存储的时候每个像素存储为8bit
- L：8位像素，表示黑和白。
- P：8位像素，使用调色板映射到其他模式。
- I：32位整型像素。
- F：32位浮点型像素。
- RGB：3x8位像素，为真彩色。
- RGBA：4x8位像素，有透明通道的真彩色。
- CMYK：4x8位像素，颜色分离。
- YCbCr：3x8位像素，彩色视频格式。

```python
plt.imshow(im.convert('L'))
```

### 图像增强

#### [滤波器](https://blog.csdn.net/FloatDreamed/article/details/79015551)
- ImageFilter模块包含许多可以与filter()方法一起使用的预定义滤波器
- BLUR：模糊滤波
- CONTOUR：轮廓滤波
- DETAIL：细节滤波
- EDGE_ENHANCE：边界增强滤波
- EDGE_ENHANCE_MORE：深度边缘增强滤波
- EMBOSS：浮雕滤波
- FIND_EDGES：寻找边界滤波（找寻图像的边界信息）
- SMOOTH：平滑滤波
- SMOOTH_MORE：深度平滑滤波
- SHARPEN：锐化滤波
- GaussianBlur：高斯模糊
- UnsharpMask：反锐化掩码滤波
- (radius：模糊半径；percent：反锐化强度（百分比）；threshold：被锐化的最小亮度)
- Kernel：卷积核滤波
- (size：核的大小；kernel：核权值序列如33的为；scale：缩放因子；offset：偏移量)
- RankFilter：排序滤波
- (size：核的大小；rank：排序序号)
- Kernel：卷积核滤波
- (size：核的大小；kernel：核权值序列如33的为；scale：缩放因子；offset：偏移量)
- MinFilter：最小值滤波器
- MedianFilter：中值滤波
- MaxFilter：最大值滤波

#### 点操作
point()方法可用于转换图像的像素值(例如图像对比度操作)

```python
source = im.split()
R,G,B = 0, 1, 2

mask = source[R].point(lambda i: i< 100 and 255)
plt.imshow(mask)
print(mask)
```

```python
out = source[G].point(lambda i: i* 0.3)
plt.imshow(out)
```

```python
source[G].paste(out,None,mask)
im = Image.merge(im.mode, source)
plt.imshow(im)
```

#### 增强
对于更高级的图像增强，您可以使用 ImageEnhance 模块中的类。从图像创建后，可以使用增强对象快速尝试不同的设置

```python
from PIL import ImageEnhance
enim = ImageEnhance.Contrast(im)
im = enim.enhance(1.9)
plt.imshow(im)
```

### 图像序列
Python图像库包含对图像序列（也称为动画格式）的一些基本支持。支持的序列格式包括FLI/FLC, GIF和一些实验格式。 TIFF文件也可以包含多个帧

当你打开一个序列文件时，PIL自动加载序列中的第一帧。您可以使用seek和tell方法在不同的帧之间移动。注意当前版本库中的大多数驱动程序只允许您寻找下一帧。要倒回文件，您可能必须重新打开它。
