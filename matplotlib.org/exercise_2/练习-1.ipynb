## 配置图题

title标题方法，也有许多可以配置的参数：
- fontsize：字体大小，默认12，也可以使用xx-small....字符串系列
- fontweight：字体粗细，或者'light'、'normal'、'medium'、'semibold'、'bold'、 'heavy'、'black'。
- fontstyle： 字体类型，或者'normal'、'italic'、'oblique'。
- verticalalignment：垂直对齐方式 ，或者'center'、'top'、'bottom'、'baseline'
- horizontalalignment：水平对齐方式，可选参数：‘left’、‘right’、‘center’
- rotation：旋转角度
- alpha： 透明度，参数值0至1之间
- backgroundcolor： 背景颜色
- bbox：给标题增加外框 ，常用参数如下：
    - boxstyle：方框外形
    - facecolor：(简写fc)背景颜色
    - edgecolor：(简写ec)边框线条颜色
    - edgewidth:边框线条大小

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
# 采用行列索引获取单个 axes
axes[0,1].set_title('This is B', fontsize=20,fontweight='bold')
# rotation旋转度数
axes[0,0].set_title('This is A',verticalalignment='center', rotation=90)

# 不指定则默认最后一个
plt.title('Title test')
plt.plot(3,2, 'gd')

axes[1,0].set_title('This is C',bbox=dict(boxstyle='round,pad=0.5',fc='pink',ec='y',lw=2))
```

### 常规使用

```python
plt.plot(x,x,label='x-x')
plt.legend(loc=2, fontsize=20)

plt.figure(frameon=False)
plt.plot(x,np.sin(x),label='x-sin')
plt.plot(x,np.cos(x),label='x-cos')
plt.legend(frameon=False,ncol=2)
```

### 后续添加图例
legend(self, parent, handles, labels,...)

parent : `~matplotlib.axes.Axes` or `.Figure`
The artist that contains the legend.

handles : list of `.Artist`
A list of Artists (lines, patches) to be added to the legend.

labels : list of str
A list of labels to show next to the artists. The length of handles
and labels should be the same. If they are not, they are truncated
to the smaller of both lengths

```python
lines=[]
style=['-', '--','-.',':']
x = np.linspace(0,10,1000)

for i in range(4):
    lines += plt.plot(x,np.sin(x-i*np.pi/2), style[i])
    print(type(lines[i]))
# 让轴xy等比例
plt.axis('equal')
# 生成第一个图例
# legend(线实例、名字列表，设置)
leg = plt.legend(lines[:2],['LineA','LineB'],loc=1, frameon=False)

# 生成第二个图例，默认会覆盖第一个
# 当前坐标轴，使用它的 add_artist方法将第一个图例重新画上去
# plt.gca().add_artist()方法(或者ax.add_artist())
leg2 = plt.legend(lines[2:],['lineC','LineD'], loc=3)
plt.gca().add_artist(leg)
```

### 箭头
plt.arrow()方法可以实现箭头的功能，但是由于它生成的是SVG向量对象，会随着图形分辨率的变化而变化，这有可能给我们带来困扰。所以，我们一般使用plt.annotate()方法来实现箭头和注释的功能

```python
fig, ax = plt.subplots()
x = np.linspace(0, 20, 100)
ax.plot(x, np.sin(x))
ax.axis('equal')

ax.annotate('local maximum', xy=(6.28,1), xytext=(10,4), arrowprops=dict(facecolor='y',shrink=0.05))
```

### 箭头

```python
fig = plt.figure(1, figsize=(8,5))
ax = fig.add_subplot(111,autoscale_on=False, xlim=(-1, 5), ylim=(-4, 3))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=3)

ax.annotate('straight',
            xy=(0, 1), xycoords='data',
            xytext=(-50, 30), textcoords='offset points',
            arrowprops=dict(arrowstyle="->"))
ax.annotate('arc3,\nrad 0.2',
            xy=(0.5, -1), xycoords='data',
            xytext=(-80, -60), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.2"))
ax.annotate('angle,\nangle 90',
            xy=(2., 1), xycoords='data',
            xytext=(-70, 30), textcoords='offset points',
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="angle,angleA=0,angleB=90,rad=10"))
```
