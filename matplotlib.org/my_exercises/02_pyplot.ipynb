#### matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLA

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.ylabel('some numbers')
```

```python
# plot()是一个通用命令，可以接受任意数量的参数
plt.plot([1,2,3,4],[1,4,9,16])
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
# axis(xmin, xmax, ymin, ymax)
plt.axis([0,6,0,20])
```

```python
# you will use numpy arrays in plt
import numpy as np
t = np.arange(0,5,0.2)
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
```

#### Plotting with keyword strings

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0,50,50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
```
