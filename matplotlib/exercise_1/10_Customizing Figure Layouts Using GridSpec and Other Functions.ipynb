#### How to create grid-shaped combinations of axes.
- subplots（）
- GridSpec()
- SubplotSpec()
- subplot2grid()

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# Using subplots() is quite simple. 
# It returns a Figure instance and an array of Axes objects.

fig1, f1axes = plt.subplots(2,2)
```

```python
fig1, f1axes = plt.subplots(2,2,constrained_layout=True)
```

```python
# The power of gridspec comes in being able to create subplots that span rows and columns. 
fig2 = plt.figure(constrained_layout=True)
spec2 = gridspec.GridSpec(2,2,figure=fig2)
f2_ax1 = fig2.add_subplot(spec2[0,0])
f2_ax2 = fig2.add_subplot(spec2[1,1])
```

```python
fig3 = plt.figure(constrained_layout=True)

# spec3 = gridspec.GridSpec(3,3,fig3)

spec3_ax1 = fig3.add_subplot(spec3[1,:])
spec3_ax1 = fig3.add_subplot(spec3[2,2])
```
