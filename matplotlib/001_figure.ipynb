#### figure:幕布/底图
- figure 是一个大窗口
- plt.figure()用于创建一个窗口，要将图画在多个窗口，可以多次调用
- fig多种可以有多个axis

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,10)
y1 = 2*x + 1
y2 = x**2
plt.figure(figsize=[5,3])
plt.plot(x,y2)
# 要将图画在两个窗口，就调用两次
# plt.figure()
plt.plot(x,y1,color='red',linewidth=1,linestyle='--')
```

#### 坐标轴

```python
plt.xlim(-5,5)
plt.ylim(-10,10)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
```

#### 图例

```python
# 要用handles引用的时候，后面注意加逗号
l1,= plt.plot(x,y2,label='up')
l2,= plt.plot(x,y1,color='red',linewidth=1,linestyle='--',label='down')
# 要显示图例需要输入plt.legend(handles,labels,loc)
plt.legend()
ax = plt.gca()
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('right')
ax.spines['bottom'].set_position(('axes', 0.4))
ax.spines['right'].set_position(('data',0))
```

```python
# 定制图例的位置
l1,= plt.plot(x,y2,label='up')
l2,= plt.plot(x,y1,color='red',linewidth=1,linestyle='--',label='down')
plt.legend(handles=[l1,l2],labels=['aaa','bbb'],loc=9)
```

```python
x = np.linspace(-3,3,50)
y = 0.1*x

plt.figure()
plt.plot(x,y,linewidth=10)
plt.ylim(-2,2)

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))


for label in ax.get_xticklabels()+ ax.get_yticklabels():
    label.set_fontsize(12)
    label.set_bbox(dict(facecolor='white',edgecolor='black',alpha=0.7))
    
```

#### 散点图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X) # for color value

plt.scatter(X,Y,s = 75, c = T,alpha=0.5)
# plt.scatter(np.arange(10),np.arange(10))

plt.xlim((-1.5,1.5))
plt.ylim((-1.5,1.5))
plt.xticks(())
plt.yticks(())
```

#### 柱状图

```python
n = 12
x = np.arange(n)
y1 = (1-x/float(n))*np.random.uniform(0.5,1,n)
y2 = (1-x/float(n))*np.random.uniform(0.5,1,n)

plt.bar(x,y1,facecolor='#9999ff',edgecolor='white')
plt.bar(x,-y2,facecolor='#ff9999',edgecolor='white')

for x,y in zip(x,y1):
    plt.text(x,y+0.05,'%.2f'%y,ha='center',va='bottom',size=9)


plt.xlim(-.5,n)
plt.xticks(())
plt.ylim(-1.25,1.25)
plt.yticks(())
```

#### 3D图形

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4,4,0.25)
Y = np.arange(-4,4,0.25)
X,Y = np.meshgrid(X,Y)
R = np.sqrt(x**2 + y**2)
Z = np.sin(R)

ax.plot_surface(X,Y,Z,rstrie=1,cstride=1,cmap=plt.get_cmap('rainbow'))




```
