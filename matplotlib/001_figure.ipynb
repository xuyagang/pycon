#### figure:幕布/底图
- figure 是一个大窗口
- plt.figure()用于创建一个窗口，要将图画在多个窗口，可以多次调用
- fig多种可以有多个axis

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,10)
y1 = 2*x + 1
y2 = x**2
plt.figure(figsize=[5,3])
plt.plot(x,y2)
# 要将图画在两个窗口，就调用两次
# plt.figure()
plt.plot(x,y1,color='red',linewidth=1,linestyle='--')
```

#### 坐标轴

```python
plt.xlim(-5,5)
plt.ylim(-10,10)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
```

#### ticks

```python
new_tick = np.linspace(-2,4,5)
plt.xticks(new_tick)
```

```python
plt.xticks([-1,0,1.5,3],['really bad','bad','normal','good','really good'])
```

```python
# 改变ytick的字体
plt.xticks([-1,0,1.5,3],['$really bad$','$bad$','$normal$','$good$','$really good$'])
```

```python
# 插入字符
plt.xticks([-1,0,1.5,3],['$really bad$','$bad\ \\alpha$','$normal$','$good$','$really good$'])
```

```python
# 修改坐标轴的位置
# gca = get current axes
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.spines['bottom'].set_position(('data',0.5))
ax.spines['left'].set_position(('data',0.5))

# outward,ax
```

#### 图例

```python
# 要用handles引用的时候，后面注意加逗号
l1,= plt.plot(x,y2,label='up')
l2,= plt.plot(x,y1,color='red',linewidth=1,linestyle='--',label='down')
# 要显示图例需要输入plt.legend(handles,labels,loc)
plt.legend()
```

```python
# 定制图例的位置
l1,= plt.plot(x,y2,label='up')
l2,= plt.plot(x,y1,color='red',linewidth=1,linestyle='--',label='down')
plt.legend(handles=[l1,l2],labels=['aaa','bbb'],loc=9)
```
