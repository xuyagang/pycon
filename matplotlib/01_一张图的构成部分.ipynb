```python
import matplotlib.pyplot as plt
# %matplotlib inline
import numpy as np
```

```python
data = np.arange(10)
plt.plot(data)
print(data)
```

#### 线条宽度

```python
plt.plot(data,lw=5)
```

#### 支持颜色

#### 线型选择

```python
plt.plot(data, color = 'r', linestyle = ':')
```

#### marker 选择

```python
plt.plot(data,color='r',marker='o')
```

#### 添加图例

```python
# 图例必须放在图之后设置
plt.plot(data,color='g',marker='.')
plt.legend('y')
```

#### 添加标题

```python
# 中文和负号的正常显示
# plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
# plt.rcParams['axes.unicode_minus'] = False


plt.plot(data)
plt.legend('x')
plt.title('简单可视化')

#用来正常显示中文标签
plt.rcParams['font.sans-serif'] = ['SimHei']
#用来正常显示负号
plt.rcParams['axes.unicode_minus'] = False
```

#### 添加水平和竖直轴文字

```python
plt.plot(data)
# 显示图例
plt.legend('y')
plt.title('简单可视化')
# 中文和负号的正常显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 添加水平轴和竖直轴文字
plt.xlabel('x轴')
plt.ylabel('y轴')
```

#### 设置绘图风格

```python
plt.style.use('ggplot')
data = np.arange(10)
data
plt.plot(data)

# 设置图形显示风格

# 显示图例
plt.legend('y')
plt.title('It is a 测试')
# 中文负号正常显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

#### 添加标注文字
annotate

```python
plt.style.use('ggplot')
data = np.arange(10)
data
plt.plot(data)

# 设置图形显示风格

# 显示图例
plt.legend('y')
plt.title('It is a 测试')
# 中文负号正常显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 水平轴和竖直文字
plt.xlabel('我是x')
plt.ylabel('我是y')

# 注释文字
# xy 是文字起始点
plt.annotate('我是标注文字', xy=(3,5))


# xytext 是注释起始点
# xy 是箭头终点
plt.annotate('带箭头标注',xy=(6,6),xytext=(3,8),arrowprops=dict(facecolor='black'),)
```
