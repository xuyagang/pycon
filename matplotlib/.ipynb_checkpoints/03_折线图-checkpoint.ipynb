##### 实战案例
- 散点图 （利用分布形态反映变量统计关系）
- 折线图 （显示随变量而变化的连续数据）
- 直方图 （是数值数据分布的精确图形表示）
- 条形图 （显示各个项目之间的比较情况）
- 饼图 （显示每一数值相对于总数值是大小）
- 箱线图 （用作显示一组数据分散情况的统计图）

```python
import pandas as pd
import matplotlib.pyplot as plt

# 设置中文正负号正常显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] =  False

# 读取要绘图的数据
df = pd.read_excel(r'C:\Users\Dell\Desktop\dataset\test.xls',)
```

```python
df
```

```python
# 删除无效
df = df.dropna(axis=0)
df
```

```python
# 开始绘图
# 图框大小
fig = plt.figure(figsize=(8,4))
plt.plot(df['指标'],df['2017年'])
# 添加标题和坐标轴标签
plt.title('data analysis test')
plt.xlabel('类别')
plt.ylabel('费用')

fig.autofmt_xdate(rotation=45)
```

```python
# 设置图框大小
fig = plt.figure(figsize=(9,3))
# 设置绘图风格
plt.style.use('seaborn')
# 设置中文编码和负号的正常显示
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
# 绘图
plt.plot(df['指标'],df['2015年'])
plt.plot(df['2014年'],
        marker = 'o',
        markersize= 5,
        markerfacecolor = 'mediumpurple'
        )
# 添加标题
plt.title('2016年数据')
# 添加坐标轴标签
plt.xlabel('类别')
plt.ylabel('费用')
# 横坐标文字旋转
fig.autofmt_xdate(rotation= 45)
```
