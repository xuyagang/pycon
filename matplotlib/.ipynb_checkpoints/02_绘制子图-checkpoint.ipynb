```python
import numpy as np
import matplotlib.pyplot as plt
# 在jupyter notebook 页面显示绘制图
%matplotlib inline
```

```python
# 分成2*2,占用第一个，即第一行第一列的子图
plt.subplot(221)
# 分成2*2,占用第二个，即第一行第二列的子图
plt.subplot(222)
# 分成2*2,占用第二个，即第一行第二列的子图
plt.subplot(212)
```

```python
def f(t):
#     np.exp(-t)—— e 的 -t 次方
    return np.exp(-t)* np.cos(2 * np.pi * t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

ax1 = plt.subplot(221)
ax1.plot(t1, f(t1), 'bo')

ax2 = plt.subplot(222)
ax2.plot(t2, np.cos(2 * np.pi * t2),'r-')

ax3 = plt.subplot(212)
ax3.plot([1,2,3,4], [1,4,9,16])
```

```python
# 中文和负号正常显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


def f(t):
#     np.exp(-t)—— e 的 -t 次方
    return np.exp(-t)* np.cos(2 * np.pi * t)

t1 = np.arange(0, 5, 0.1)
t2 = np.arange(0, 5, 0.02)

ax1 = plt.subplot(221)
ax1.set_title('这是子图1')
ax1.plot(t1, f(t1), 'bo')

ax2 = plt.subplot(222)
ax2.set_title('这是子图2')
ax2.plot(t2, np.cos(2 * np.pi * t2),'r-')

ax3 = plt.subplot(212)
ax3.set_xlabel('年龄')
ax3.set_ylabel('收入')
```
