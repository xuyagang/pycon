```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# dataframe 的默认展示是按列进行，将每列分为一类
df = pd.DataFrame(np.random.randn(10,4),
                  index=pd.date_range('2019/01/01', periods=10),
                 columns=list('ABCD'))
print(df)
```

```python
df.plot()
```

##### 如果索引由日期组成，则调用gct().autofmt_xdate()来格式化x轴

#### 条形图

```python
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','c'])
df.plot.bar()
```

```python
# 要生成一个堆积条形图，通过指定:stacked=True
df.plot.bar(stacked=True)
```

```python
# 要获得水平条形图，使用barh()方法
df.plot.barh()
```

```python
df.plot.barh(stacked=True)
```

#### 直方图
使用plot.hist()方法绘制直方图

```python
df = pd.DataFrame({'a':np.random.randn(1000)+1, 'b':np.random.randn(1000),
                  'c':np.random.randn(1000)-1}, columns=['a','b','c'])
```

```python
df.plot.hist(bins=20,alpha=0.3)
```

```python
# 要为每列绘制不同的直方图，请使用以下代码
df.hist(bins=20)
```

#### 箱形图
Boxplot可以绘制调用Series.box.plot()和DataFrame.box.plot()或DataFrame.boxplot()来可视化每列中值的分布

```python
df = pd.DataFrame(np.random.rand(10,5), columns=['A', 'B', 'C', 'D', 'E'])
df.plot.box()
```

```python
df.boxplot()
```

#### 区域块图形
可以使用Series.plot.area()或DataFrame.plot.area()方法创建区域图形

```python
df = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','c'])
df.plot.area(alpha=0.5)
```

```python
df = pd.DataFrame(np.random.rand(50,4),columns=['a','b','c','d'])
df.plot.scatter(x= 'a',y='c')
```

```python
df = pd.DataFrame(3 * np.random.rand(4), index=['a', 'b', 'c', 'd'], columns=['x'])
df.plot.pie(subplots=True)
```
